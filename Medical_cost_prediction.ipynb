## **Data load**

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/Deepakthota91/Data/refs/heads/main/Medical_Cost_Prediction.csv')
df
```

# **Data Info**

```python
print("Dataset Info:")
print(df.info())
df['BMI'].fillna(df['BMI'].mean(), inplace=True)
df['Annual_Income'].fillna(df['Annual_Income'].median(), inplace=True)
df.dropna(subset=['Medical_Expenses'], inplace=True)
print("Dataset Info:")
print(df.info())
```

# **Label Encodeing**

```python
from sklearn.preprocessing import LabelEncoder


label_encoder = LabelEncoder()

df['gender_encoded'] = label_encoder.fit_transform(df['Gender'])

print("\nDataset after Label Encoding:")
df
```

# **Minmax Scaling**



```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['scaled_Annual_income'] = scaler.fit_transform(df[['Annual_Income']])
df['scaled_Medical_Expenses'] = scaler.fit_transform(df[['Medical_Expenses']])
df
```

## **Finding dependent and independent values**

```python
y = df['scaled_Medical_Expenses']
y
```

```python
x = df.drop(['Gender','Patient_ID','Medical_Expenses','Annual_Income'],axis=1)
x
```

## **Splitting the Dataset**

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=53)

x_train
```

## **Linear Regression and training**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

# **prediction of y_test**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

# **Accuracy Testing**

```python
from sklearn.metrics import mean_squared_error,r2_score

lr_train_mse = mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)
lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
lr_results= pd.DataFrame(['Linear regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns=['Method','Training MSE','Training R2','Testing MSE','Testing R2']
lr_results
```

# **Pickle**

```python
import pickle
with open('Medical_cost_predict.pkl', 'wb') as file:
  pickle.dump(lr,file)
print("Linear Regression saved as 'Medical_cost_predict.pkl'")
```

## **Predictions with .pkl**

```python
with open('Medical_cost_predict.pkl', 'rb') as file:
    loaded_model = pickle.load(file)

predictions = loaded_model.predict(x_test)
original_max=df['Medical_Expenses'].min()
original_min=df['Medical_Expenses'].max()
upscaled_predictions = predictions * (original_max - original_min) + original_min
print("Predictions:", upscaled_predictions)
```

```python
from flask_ngrok import run_with_ngrok
from flask import Flask, request, render_template
import pickle
import numpy as np

with open('Medical_cost_predict.pkl', 'rb') as file:
  model = pickle.load(file)
app = Flask(__name__)
run_with_ngrok(app)
@app.route('/')
def home():
    return render_template('Medical_bill_prediction.html')

@app.route('/predict', methods=['POST'])
def predict():
    age = request.form['age']
    bmi = request.form['bmi']
    chronic_conditions = request.form['chronicConditions']
    doctors_visited = request.form['doctorsVisited']
    surgical_procedures = request.form['surgicalProcedures']
    hospital_stays = request.form['hospitalStays']
    medication_prescribed = request.form['medicationPrescribed']
    gender_encoded = request.form['genderEncoded']
    annual_income = request.form['annualIncome']


    input_data = np.array([[age, bmi, chronic_conditions, doctors_visited,
                            surgical_procedures, hospital_stays,
                            medication_prescribed, gender_encoded,
                            annual_income]])


    prediction = model.predict(input_data)

    return render_template('Medical_bill_prediction.html', prediction_text=f'Estimated Hospital Expense: ${prediction[0]:.2f}')

if __name__ == "__main__":
    app.run()

```
